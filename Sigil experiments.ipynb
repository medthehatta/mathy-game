```python
%matplotlib inline
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import sample
from iplotting import *
from plotly.offline import iplot as plotly_iplot, init_notebook_mode
init_notebook_mode(connected=True)
import plotly.graph_objs as go
from functools import reduce
import octonion
from importlib import reload
```

```python
R = octonion.oR
I = octonion.oI
J = octonion.oJ
K = octonion.oK
L = octonion.oL
IL = octonion.oIL
JL = octonion.oJL
KL = octonion.oKL
```

```python
def flatten(lst):
    return sum(lst, [])
```

```python
proj_ri = proj('ri')
proj_rj = proj('rj')
proj_rk = proj('rk')
proj_ij = proj('ij')
proj_jk = proj('jk')
proj_rik = proj('rik')
proj_rjk = proj('rjk')
proj_ijk = proj('ijk')
```

```python
N = 3
viewN = 3

x = 1 + np.tan(np.pi/N)*I

def rotor(angle, q):
    return np.cos(angle)*R + np.sin(angle)*(q.I*I + q.J*J + q.K*K)

rotork = rotor(np.pi/3, K)
rotork = rotork/rotork.norm()

all_starts = [x, rotork]

xss = [all_starts] + [
    sample.iterated_products(all_starts, n=n)
    for n in range(1, viewN + 1)
]

plot_generations(xss, proj('ri'))
plot_generations(xss, proj('rj'))
plot_generations(xss, proj('ij'))
plot_generations(xss, proj('rij'))
```
